```python
import pandas as pd

# Loading the training and test datasets
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')

# Previewing the datasets
print(train_data.head())
print(test_data.head())
```

```python
# Selecting features and target
features = train_data[['GrLivArea', 'BedroomAbvGr', 'FullBath']]
target = train_data['SalePrice']

# Handling missing values if necessary
features = features.dropna()
target = target.loc[features.index]
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

```python

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.metrics import mean_squared_error

# Defining a function to compute the cost
def compute_cost(m, b, X, y):
    predictions = m * X + b
    cost = mean_squared_error(y, predictions)
    return cost

# Generating a range of values for m and b
m_values = np.linspace(-1, 1, 100)
b_values = np.linspace(-200000, 200000, 100)
M, B = np.meshgrid(m_values, b_values)
costs = np.zeros(M.shape)

# Computing the cost for each combination of m and b
for i in range(M.shape[0]):
    for j in range(M.shape[1]):
        costs[i, j] = compute_cost(M[i, j], B[i, j], features['GrLivArea'], target)

# Plotting the cost function using a 3D surface plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(M, B, costs, cmap='viridis', edgecolor='none')

# Adding labels and title
ax.set_xlabel('Slope (m)')
ax.set_ylabel('Intercept (b)')
ax.set_zlabel('Cost')
ax.set_title('3D Surface Plot of Cost Function for Linear Regression')

# Adding a color bar to show cost values
fig.colorbar(surf, shrink=0.5, aspect=5, label='Cost')

plt.show()
```
