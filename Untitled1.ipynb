```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler

# 1. Loading and Preparing the Data
df = pd.read_csv('train.csv')
features = df[['GrLivArea', 'BedroomAbvGr', 'FullBath']].fillna(0).values
target = df['SalePrice'].values

# Normalizing features and target
scaler_X = StandardScaler()
scaler_y = StandardScaler()

features_normalized = scaler_X.fit_transform(features)
target_normalized = scaler_y.fit_transform(target.reshape(-1, 1)).flatten()

# Adding a bias column (intercept) to the features
X = np.hstack((np.ones((features_normalized.shape[0], 1)), features_normalized))
y = target_normalized

# 2. Defining the Linear Regression Model
def predict(X, weights):
    return X.dot(weights)

def compute_cost(X, y, weights):
    predictions = predict(X, weights)
    cost = np.mean((y - predictions) ** 2)  # Calculating Mean Squared Error (MSE)
    return cost

def gradient_descent(X, y, weights, learning_rate, num_iterations):
    m = len(y)
    cost_history = np.zeros(num_iterations)
    for i in range(num_iterations):
        predictions = predict(X, weights)
        errors = predictions - y
        gradients = (2/m) * X.T.dot(errors)
        weights -= learning_rate * gradients
        cost_history[i] = compute_cost(X, y, weights)  # Tracking cost over iterations
    return weights, cost_history

# Initializing parameters
num_features = X.shape[1]
weights = np.zeros(num_features)
learning_rate = 0.001
num_iterations = 1000

# Performing Gradient Descent
weights, cost_history = gradient_descent(X, y, weights, learning_rate, num_iterations)

# 3. Visualizing the Cost Function
# Generating a range of values for the first feature and bias term
m_values = np.linspace(-1, 1, 100)
b_values = np.linspace(-2, 2, 100)
M, B = np.meshgrid(m_values, b_values)
costs = np.zeros(M.shape)

# Computing the cost for each combination of the first feature and bias term
for i in range(M.shape[0]):
    for j in range(M.shape[1]):
        weights_temp = np.zeros(num_features)
        weights_temp[0] = B[i, j]  # Bias term
        weights_temp[1] = M[i, j]  # First feature coefficient
        # For simplicity, setting other feature coefficients to 0
        costs[i, j] = compute_cost(X, y, weights_temp)

# Plotting the cost function using a 3D surface plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(M, B, costs, cmap='viridis', edgecolor='none')

# Adding labels and title to the plot
ax.set_xlabel('Intercept (b)')
ax.set_ylabel('Feature 1 Coefficient (m1)')
ax.set_zlabel('Cost')
ax.set_title('3D Surface Plot of Cost Function for Linear Regression')

# Adding a color bar to show cost values
fig.colorbar(surf, shrink=0.5, aspect=5, label='Cost')

plt.show()

# 4. Making Predictions and Evaluating the Model
y_pred_normalized = predict(X, weights)
y_pred = scaler_y.inverse_transform(y_pred_normalized.reshape(-1, 1)).flatten()  # Inverse transform to original scale
mse = compute_cost(X, target_normalized, weights)  # Calculating Mean Squared Error
print(f'Mean Squared Error: {mse}')

# Plotting Actual vs. Predicted values
plt.figure(figsize=(10, 6))
plt.scatter(target, y_pred, color='blue', label='Predicted vs Actual')
plt.plot([target.min(), target.max()], [target.min(), target.max()], color='red', linestyle='--', label='Perfect Fit')
plt.xlabel('Actual Sale Price')
plt.ylabel('Predicted Sale Price')
plt.title('Actual vs. Predicted Sale Price')
plt.legend()
plt.show()
```
